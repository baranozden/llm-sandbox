```python
from abc import ABC, abstractmethod
from bs4 import BeautifulSoup
from dataclasses import dataclass, field
import requests
```

```python
def get_request(url):
    """Sends a get request to the url given in the initialization"""
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36',
        'Accept': 'text/html,text/plain,application/xhtml+xml,application/xml,application/_json;q=0.9,image/webp,image/apng,*/*;q=0.8',
        'Accept-Charset': 'Windows-1252,utf-8;q=0.7,*;q=0.3',
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'en-US,en;q=0.8;q=0.5',
        'Connection': 'keep-alive'
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        content = BeautifulSoup(response.content)
        return content
    except requests.exceptions.HTTPError as e:
        print(e)

def parse_content(self, parser) -> tuple:
    """
    Parses html content and returns final result

    :param parser: ParsingStrategy object, takes appropriate parser
    :return: tuple, final parsed output as product, image, price, currency
    """
    self.get_request()
    return parser.parse_response(self.content)
```
